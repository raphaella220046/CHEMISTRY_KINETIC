Após diversos problemas com o modelo da simulação atomística 2D, resolvemos trocar para o projeto 2, que consistia em usar dados experimentais, retirados de FOUSTOUKOS, Dionysis I.; HOUGHTON, Jennifer L.; SEYFRIED, William E.; et al. Kinetics of H2–O2–H2O redox equilibria and formation of metastable H2O2 under low temperature hydrothermal conditions. Geochimica et Cosmochimica Acta, v. 75, n. 6, p. 1594–1607, 2011. Disponível em: <https://linkinghub.elsevier.com/retrieve/pii/S0016703710006940>. Acesso em: 18 out. 2023.

```python
import numpy as np
import matplotlib.pyplot as plt

# Concentrações iniciais
H2_0 = 50
O2_0 = 50
H2O_0 = 0
O_0 = 0

# Constante de velocidade
k = 0.000535

# Tempo inicial e final da simulação
t0 = 0
tf = 60

# Número de pontos na simulação
n = 1000

# Intervalo de tempo
dt = (tf - t0) / n

# Vetores para armazenar as concentrações e o tempo
H2 = np.zeros(n)
O2 = np.zeros(n)
H2O = np.zeros(n)
O = np.zeros(n)
t = np.zeros(n)

# Condições iniciais
H2[0] = H2_0
O2[0] = O2_0
H2O[0] = H2O_0
O[0] = O_0
t[0] = t0

# Loop para simular a reação
for i in range(1, n):
    # Cálculo das concentrações no instante i
    v = k * H2[i-1]**2 * O2[i-1]
    H2[i] = H2[i-1] - 2 * v * dt
    O2[i] = O2[i-1] - v * dt
    H2O[i] = H2O[i-1] + v * dt
    O[i] = O[i-1] + v * dt
    t[i] = t[i-1] + dt

# Plot do gráfico
plt.title('Gráfico da []xt, a T = 523,15 Kelvin, (k = 0.000535)')
plt.plot(t, H2, label='H2')
plt.plot(t, O2, label='O2')
plt.plot(t, H2O, label='H2O')
plt.plot(t, O, label='O', linestyle='dashed')
plt.xlabel('Tempo (s)')
plt.ylabel('Concentração (mols)')
plt.legend()
plt.show()
```
